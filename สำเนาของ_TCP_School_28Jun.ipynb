ในทางวิทยาศาสตร์ เราสามารถสร้างนิยามเชิงปริมาณของการเปลี่ยนแปลงให้อยู่ในรูปของ**ระบบพลวัต (Dynamical System)** ซึ่งเป็นการพิจารณาการเปลี่ยนแปลงของตัวแปร $x$ ชนิดหนึ่งที่เป็นฟังก์ชั่นของตัวแปรต้นคือเวลา $t$ โดยที่อัตราการเปลี่ยนแปลง $\mathrm{d}x/\mathrm{d}t$ อาจจะขึ้นกับพารามิเตอร์ต่างๆของปัจจัย หรืออาจจะขึ้นกับตัวแปรของระบบที่มีความเกี่ยวพันกันในระบบสมการเชิงอนุพันธ์ (systems of differential equations) เช่น ในกรณีที่ระบบ 1 และ 2 มีความเกี่ยวพันกันเราอาจบอกได้ว่า $\mathrm{d}x_1/\mathrm{d}t$ มีค่าขึ้นอยู่กับตัวแปร $x_2$ และในขณะเดียวกัน $\mathrm{d}x_2/\mathrm{d}t$ มีค่าขึ้นอยู่กับตัวแปร $x_1$ เช่นกัน ในระบบพลวัตที่ประกอบด้วยหลายตัวแปร เช่น ระบบของห่วงโซ่การเกิดปฏิกิริยาเคมีในสิ่งมีชีวิต หรือการเปลี่ยนแปลงของราคาหุ้นที่ขึ้นอยู่กับปัจจัยจำนวนมาก การแก้ระบบสมการเพื่อให้ได้มาซึ่งฟังก์ชั่น $x_i (t)$ สำหรับแต่ละตัวแปร $i$ ไม่สามารถทำได้โดยง่ายด้วยวิธีการวิเคราะห์สมการ ระเบียบวิธีเชิงตัวเลข (numerical methods) จึงเข้ามาเป็นตัวช่วยในการได้มาซึ่งข้อมูลการเปลี่ยนแปลงที่สมบูรณ์โดยจะใช้พลังในการคำนวณของคอมพิวเตอร์ในยุคปัจจุบันที่สามารถทำงานได้อย่างรวดเร็วและยังคงมีพัฒนาการอย่างต่อเนื่อง ซึ่งแม้ว่าการคำนวณโดยใช้การประมาณค่าเชิงตัวเลขยังคงมีข้อจำกัดที่ไม่สามารถนำไปสู่ผลเฉลยแม่นตรง (exact solution) ได้เลย แต่ก็ยังนับว่าเป็นวิธีที่มีประสิทธิภาพที่สุดในการแก้ปัญหาเชิงพลวัตที่มีความซับซ้อน

**2. ระเบียบวิธีของออยเลอร์ (Euler Method)**

เป็นระเบียบวิธีพื้นฐานในการแก้สมการเชิงอนุพันธ์อันดับหนึ่งในรูปของ $x'(t)=f(t,x(t))$ ซึ่งจะทำการประมาณค่าเชิงเส้นของส่วนอัตราการเปลี่ยนแปลงทางด้านซ้ายมือของสมการ ในกรณีที่ตัวแปรต้นของสมการเชิงอนุพันธ์คือเวลา เราสามารถประมาณอัตราการเปลี่ยนแปลงในช่วงเวลา $\Delta t$ สั้นๆ ได้เป็น

$~$

\begin{equation}
\dfrac{\mathrm{d}x}{\mathrm{d}t} \approx \dfrac{\Delta x}{\Delta t} = f(t,x(t)) \tag{1}
\end{equation}

$~$

และเมื่อใช้วิธีการ time discretization แบ่งช่วงเวลาเป็นขั้นเวลา (timestep) สั้นๆ ช่วงละ ${\Delta t}$ ทั้งหมด $N$ ช่วง เราสามารถที่จะแก้หาค่า $x$ ณ timestep ที่ $i+1$ หรือ $x_{i+1}$ ได้ก็ต่อเมื่อเราทราบค่า $x_{i}$ ณ timestep ก่อนหน้าหรือ $i$ ด้วยวิธีการคำนวณดังนี้

$~$

\begin{equation}
\dfrac{\Delta x}{\Delta t} = \dfrac{x_{i+1}-x_i}{\Delta t} = f(t,x_i(t))
\end{equation}
นั่นคือ
\begin{equation}
x_{i+1} = x_{i} + f(t,x_i(t)) ~ \Delta t \tag{2}
\end{equation} 

$~$

ซึ่งเมื่อทำการกำหนดเงื่อนไขเริ่มต้น หรือทราบค่าของ $x_0$ จำทำให้เราสามารถหาค่าของ $x_1$ $x_2$ ไปจนถึง $x_N$ ได้โดยอาศัยการทำงานแบบวนลูป ตามตัวอย่างการคำนวณการสลายตัวของสารกัมมันตรังสีที่อัตราการสลายตัวของธาตุกัมมันตรังสีจำนวน $n$ อะตอม ซึ่งอัตราการเปลี่ยนแปลงของจำนวนอะตอมธาตุกัมมันตรังสี $\mathrm{d}n/\mathrm{d}t$ **ที่มีค่าลบ** จะขึ้นอยู่กับจำนวนอะตอมดังสมการ

$~$

\begin{equation}
\dfrac{\mathrm{d}n}{\mathrm{d}t} = -\lambda n \tag{3}
\end{equation} 

$~$

โดยที่ $\lambda$ คือค่าคงที่ของการสลายตัวมีหน่วยเดียวกับความถี่คือเฮิร์ตซ์ เมื่อลองพิจารณาช่วงเวลาในการเกิดการสลายตัวให้เป็นขั้นเวลา (timestep) เล็กๆขนาด $\Delta t = 0.1$ วินาที เราสามารถทำนายการสลายตัวของกัมมันตรังสี ณ timestep ที่ $i+1$ ต่างๆ ได้จาก

$~$

\begin{equation}
n_{i+1} = n_{i} - \lambda n_i \Delta t \tag{4}
\end{equation}

$~$

ซึ่งถ้าเงื่อนไขตั้งต้นของการสลายตัวคืออะตอมมีจำนวน $n_0 = 10000$ อะตอม ที่เวลา $t=0$ และพารามิเตอร์คือค่าคงที่ในการสลายตัวมีค่าเป็น $\lambda=3$ เฮิร์ตซ์ เราสามารถใช้วิธีการของออยเลอร์ในการทำนายจำนวนอะตอมคงเหลือที่เวลา $t$ ค่าต่างๆได้ดังโปรแกรม


````
import numpy as np
import matplotlib.pyplot as plt

# run parameters
N = 50 # number of timesteps
n0 = 10000 # initial number of atoms
lmbda = 3 # decay constant
delta_t = 0.1 # timesteps size

# array to contain time t at timestep 0 to N
t = np.arange( 0, (N+1)*delta_t, delta_t )

# array to contain number of atoms at timestep 0 to N
n = np.zeros( N+1 )

# initial condition: n at timestep 0 = initial number of atoms
n[0] = n0

# Euler method
for i in range( 0, N ):
  n[i+1] = n[i] - lmbda*n[i]*delta_t # from equation (4)

# plotting
fig, ax = plt.subplots()
plt.ylim(0, n0)
plt.xlim(0, N*delta_t)
ax.set_ylabel("n")
ax.set_xlabel("t (seconds)")
ax.plot( t, n, label='Euler' ) 
ax.plot( t, n0*np.exp(-lmbda*t), label='Exact' ) 
ax.legend()
plt.show()
````

จะเห็นได้ว่าผลการคำนวณจาก Euler method มีค่าใกล้เคียงกับผลเฉลยแม่นตรง (exact solution) ในรูปแบบของ exponential decay หรือ $n(t) = n_0 \mathrm{exp}(-\lambda t)$ ที่ได้จากการวิเคราะห์สมการ แต่ยังคงมีความคลาดเคลื่อน ซึ่งความคลาดเคลื่อนดังกล่าวมีที่มาจากการประมาณค่า $\mathrm{d}x/\mathrm{d}t$ ในรูปของ $\Delta x / \Delta t$ นั่นเอง การลดความคลาดเคลื่อนดังกล่าวสามารถทำได้โดยการลดขนาดของ timestep $\Delta t$ ลงเนื่องจาก 

$~$

\begin{equation}
\lim_{\Delta t \rightarrow 0} \dfrac{\Delta x}{\Delta t} = \dfrac{\mathrm{d}x}{\mathrm{d}t} \tag{5}
\end{equation}

$~$

และโดยการใช้ระเบียบวิธีที่มีความละเอียดสูงขึ้น เช่น Verlet algorithm หรือ Runge-Kutta ซึ่ง Euler Method เองก็จัดว่าเป็นระเบียบวิธี Runge-Kutta ที่มีลำดับขั้นในความแม่นยำหรือ order ที่ต่ำที่สุด

**Exponential Growth Model**

$
\dfrac{\mathrm{d}N}{\mathrm{d}t} = rN
$




````
# code here
# run parameters
N = 50 # number of timesteps
n0 = 2 # initial number of atoms
r = 2 # decay constant
delta_t = 0.1 # timesteps size

# array to contain time t at timestep 0 to N
t = np.arange( 0, (N+1)*delta_t, delta_t )

# array to contain number of atoms at timestep 0 to N
n = np.zeros( N+1 )

# initial condition: n at timestep 0 = initial number of atoms
n[0] = n0

# Euler method
for i in range( 0, N ):
  n[i+1] = n[i] + r*n[i]*delta_t # from equation (4)

# plotting
fig, ax = plt.subplots()
plt.ylim(0, 1000)
plt.xlim(0, N*delta_t)
ax.set_ylabel("n")
ax.set_xlabel("t (seconds)")
ax.plot( t, n, label='Euler' ) 
ax.plot( t, n0*np.exp(r*t), label='Exact' ) 
ax.legend()
plt.show()
````

**Decay Model**

$
\dfrac{\mathrm{d}N}{\mathrm{d}t} = -rN
$




````
# code here
N = 50 # number of timesteps
n0 = 1000 # initial number of atoms
r = 2 # decay constant
delta_t = 0.1 # timesteps size

# array to contain time t at timestep 0 to N
t = np.arange( 0, (N+1)*delta_t, delta_t )

# array to contain number of atoms at timestep 0 to N
n = np.zeros( N+1 )

# initial condition: n at timestep 0 = initial number of atoms
n[0] = n0

# Euler method
for i in range( 0, N ):
  n[i+1] = n[i] - r*n[i]*delta_t # from equation (4)

# plotting
fig, ax = plt.subplots()
plt.ylim(0, 1000)
plt.xlim(0, N*delta_t)
ax.set_ylabel("n")
ax.set_xlabel("t (seconds)")
ax.plot( t, n, label='Euler' ) 
ax.plot( t, n0*np.exp(-r*t), label='Exact' ) 
ax.legend()
plt.show()
````

**Logistic Growth Model**

$
\dfrac{\mathrm{d}N}{\mathrm{d}t} = rN(C-N)
$




````
# code here
N = 50 # number of timesteps
n0 = 2 # initial number of atoms
r = 2 # decay constant
delta_t = 0.1 # timesteps size
c = 1

# array to contain time t at timestep 0 to N
t = np.arange( 0, (N+1)*delta_t, delta_t )

# array to contain number of atoms at timestep 0 to N
n = np.zeros( N+1 )

# initial condition: n at timestep 0 = initial number of atoms
n[0] = n0

# Euler method
for i in range( 0, N ):
  n[i+1] = n[i] + r*(n[i]*delta_t)*(c - n[i]*delta_t) # from equation (4)

# plotting
fig, ax = plt.subplots()
plt.ylim(0, 1000)
plt.xlim(0, N*delta_t)
ax.set_ylabel("n")
ax.set_xlabel("t (seconds)")
ax.plot( t, n, label='Euler' ) 
ax.plot( t, n0*np.exp(c*r*t), label='Exact' ) 
ax.legend()
plt.show()
````

**แบบจำลองพลวัตของความรัก (Dynamic Love Model)**

(อ้างอิงจาก: Liao and Ran (2007). Hopf bifurcation in love dynamical models with nonlinear couples and time delays)

นอกจากโจทย์ตัวอย่างนี้จะมีความน่าสนใจในแง่ของการพยายามอธิบายพฤติกรรมของมนุษย์ที่พบได้ทั่วไปแล้ว ยังถือว่าเป็นตัวอย่างของปัญหาในระบบเชิงพลวัตที่ประกอบด้วยความสัมพันธ์ระหว่างตัวแปรตามมากกว่า 1 ตัว ในรูปแบบของระบบสมการเชิงอนุพันธ์อีกด้วย แนวคิดในการกำหนด (formulation) ระบบสมการขึ้นมาจากความสัมพันธ์ของคนสองคนยังสามารถนำไปใช้กับการกำหนดระบบสมการของปัญหาที่มีลักษณะคล้ายกัน เช่น การเจรจาต่อรอง  (negotiation) หรือความสัมพันธ์ในเชิงของผู้ล่ากับผู้ถูกล่า (predator-prey relationship) ได้อีกด้วย

**เงื่อนไขและการกำหนดระบบสมการ**

เราจะลองมาพิจารณากรณีสมมติของโรมิโอ (Romeo; $R$) และจูเลียต (Juliet; $J$) จากในบทความอ้างอิง โดยกำหนดตัวแปร $R(t)$ เป็นความรักของโรมิโอที่มีต่อจูเลียตและ $J(t)$ เป็นความรักของจูเลียตที่มีต่อโรมิโอ ซึ่งเป็นฟังก์ชั่นของเวลา ซึ่งอัตราการเปลี่ยนแปลงของความรู้สึก $\mathrm{d}R/\mathrm{d}t$ และ $\mathrm{d}J/\mathrm{d}t$ จะขึ้นอยู่กับปัจจัยต่างๆ เมื่อพิจารณาจากคู่ของบุคคล $m$ และ $n$ ใดๆ ได้แก่ 

$~$

1. *สเน่ห์ (Appeal) ของอีกฝ่าย* :  ซึ่งรวมเอาปัจจัยทุกอย่างไว้ด้วยกัน เช่น หน้าตา รูปร่าง นิสัยใจคอต่างๆ โดยจะกำหนดให้เป็นค่าคงที่เสน่ห์ของบุคคล $n$ หรือ $A_n$ และถ้าค่าการตอบสนองต่อเสน่ห์ของบุคคล $m$ ซึ่งบ่งบอกรสนิยมของเขาว่าชอบเพศตรงข้ามโดยตัดสินจากสเน่ห์มากน้อยแค่ไหนมีค่าเป็น $r_m$ จะทำให้อัตราการเปลี่ยนแปลงความรู้สึกของ $m$ ที่มีต่อ $n$ ซึ่งเป็นผลมาจากเสน่ห์ของ $n$ คือ

\begin{equation}
\dfrac{\mathrm{d}m}{\mathrm{d}t} = + r_m A_n \tag{6}
\end{equation}

$~$

2. *ความรักที่จืดจางไปตามกาลเวลา* :  เราจะประมาณค่าอัตราการลดลงของความรู้สึกรักโดยใช้ความสัมพันธ์แบบเดียวกับการสลายตัวของธาตุกัมมันตรังสีซึ่งเป็นแบบ exponential decay หรือ $m(t) = m_0 \mathrm{exp}(-\lambda_m t)$ คืออัตราในการเปลี่ยนแปลงที่เป็นลบจะขึ้นอยู่กับปริมาณของความรัก $m(t)$ ที่ $m$ มีต่อ $n$ อยู่ ณ ปัจจุบันตามสมการ

\begin{equation}
\dfrac{\mathrm{d}m}{\mathrm{d}t} = - \lambda_m m(t) \tag{7}
\end{equation}

3. *ความรักที่ขึ้นกับการตอบสนองต่อความรู้สึกของอีกฝ่าย* : เป็นปัจจัยที่มีความซับซ้อนเพราะนอกจากจะขึ้นอยู่กับความรู้สึกและความเร็วหรือช้าในการแสดงออกของอีกฝ่ายหรือ $n$ แล้ว ยังขึ้นอยู่กับลักษณะนิสัยของฝ่ายแรกหรือ $m$ ด้วย ซึ่งถ้าหากคิดเฉพาะปัจจัยนี้อัตราการเปลี่ยนแปลงความรู้สึกของ $m$ จะเป็นไปตามสมการ

$~$

\begin{equation}
\dfrac{\mathrm{d}m}{\mathrm{d}t} = + B_m \mathrm{tanh}(~n(t-\tau_n)~) \tag{8}
\end{equation}

$~$

โดยที่ $B_m$ เป็นค่าคงที่การตอบสนองของ $m$ ที่มีต่อความรู้สึกของ $n$ ถ้า $B_m>0$ จะแสดงให้เห็นถึงลักษณะนิสัยของ $m$ ที่จะรู้สึกรักคนที่มารักตนและเกลียดคนที่ไม่ชอบตน แต่ถ้า $B_m>0$ จะแสดงให้เห็นถึงลักษณะนิสัยของ $m$ ที่อาจจะไม่ตรงไปตรงมา เช่น ตนเองอาจจะรู้สึกลำบากใจเมื่อมีคนมารัก (หรือโลกส่วนตัวสูง) แต่จะมีความรู้สึกต้องการคนที่ไม่ชอบหรือรู้สึกกับตนเองในแง่ลบ (อาจเป็นความท้าทายหรือความอยากเอาชนะ) 
 
 อีกปัจจัยหนึ่งที่อาจมีผลคือลักษณะนิสัยการแสดงออกทางความรู้สึกของ $n$ ซึ่งสะท้อนอยู่ค่าพารามิเตอร์ความล่าช้า (delay parameter; $\tau_n$) เช่น ถ้า $\tau_n=0$ คือการที่ $n$ แสดงความรู้สึกของตนเองออกไปให้ $m$ ทราบในทันทีทันใด แต่ถ้า $\tau_n=0$ คือการที่ $n$ อาจจะลังเลหรือเลือกที่จะเก็บความรู้สึกของตนเอาไว้ในระยะหนึ่งทำให้ความรู้สึกที่ $m$ รับรู้หรือ $n(t-\tau_n)$ เป็นความรู้สึกของ $n$ ในเวลาที่ย้อนหลังกลับไป

เมื่อนำปัจจัยทั้งสามมากำหนดสมการพลวัตของความรักระหว่างโรมิโอ (Romeo; $R$) และจูเลียต (Juliet; $J$) โดยที่ให้ทั้งคู่ผลัดกันเป็นบุคคล $m$ และ $n$ จากสมการ $(6)-(8)$ จะทำให้ได้ระบบสมการเชิงอนุพันธ์ออกมาดังนี้

$~$

\begin{equation}
\dfrac{\mathrm{d}R}{\mathrm{d}t} = r_R A_J - \lambda_R R(t) + B_R \mathrm{tanh}(~J(t-\tau_J)~)
\end{equation}

และ

\begin{equation}
\dfrac{\mathrm{d}J}{\mathrm{d}t} = r_J A_R - \lambda_J J(t) + B_J \mathrm{tanh}(~R(t-\tau_R)~) \tag{9}
\end{equation}

$~$

ซึ่งในการแก้ระบบสมการด้วยวิธีออยเลอร์เราก็จะทำการ distretize เวลาโดยแบ่งเป็น timestep ละ $\Delta t$ เช่นเคย ทำให้สามารถคำนวณค่าความรู้สึกของ Romeo และ Juliet ที่ timestep $i+1$ โดยใช้ข้อมูลจาก timestep $i$ ได้ดังนี้

$~$

\begin{equation}
R_{i+1} = R_i + \big( (r_R A_J - \lambda_R R_i + B_R \mathrm{tanh}(~J_{i-\tau_J/{\Delta t}}~) \big)\Delta t
\end{equation}

และ

\begin{equation}
J_{i+1} = J_i + \big( (r_J A_R - \lambda_J J_i + B_J \mathrm{tanh}(~R_{i-\tau_R/{\Delta t}}~) \big)\Delta t \tag{10}
\end{equation}

$~$

นำไปสู่การออกแบบชุดคำสั่งในการแก้ระบบสมการหา $R(t)$ และ $J(t)$ ดังแสดง

````
import numpy as np
import matplotlib.pyplot as plt

def RJ():  
  # 1 : Romeo and 2 : Juliet
  x1 = np.zeros( nstep + 1 )
  x2 = np.zeros( nstep + 1 )

  # initial conditions
  ntau1 = int(tau1/dt)
  ntau2 = int(tau2/dt)
  x1[ 0 ] = x1_0
  x2[ 0 ] = x2_0

  # Euler method
  for i in range(0, nstep):
    i1 = max( 0, i-ntau1 )
    i2 = max( 0, i-ntau2 )
    x1[i+1] = x1[i] - lambda1*x1[i]*dt + b1*np.tanh(x2[i2])*dt + r1*app2*dt
    x2[i+1] = x2[i] - lambda2*x2[i]*dt + b2*np.tanh(x1[i1])*dt + r2*app1*dt

  # plot
  fig, ax = plt.subplots()
  ax.set_ylabel("Love")
  ax.set_xlabel("time (arbitrary unit)")
  ax.plot( x1, label='Romeo' )
  ax.plot( x2, label='Juliet' )
  ax.legend()
  plt.show()

  ntau = max( ntau1, ntau2 )
  fig, ax = plt.subplots()
  ax.set_ylabel("Juliet")
  ax.set_xlabel("Romeo")
  ax.plot( x1[ntau:], x2[ntau:] )
  ax.plot( x1[ntau], x2[ntau], 'ro' )
  plt.show( )
````

ซึ่งสามารถทำการปรับเปลี่ยนค่าพารามิเตอร์แล้วรันชุดคำสั่งในฟังก์ชั่น `RJ()` ได้ในเซลล์ด้านล่าง

````
# run parameters
dt = 0.001     # Delta t
nstep = 50000  # Number of steps
# ความจืดจาง, การตอบสนองต่อความรู้สึกฝ่ายตรงข้าม, การตอบสนองต่อสเน่ห์, สเน่ห์, การเก็บความรู้สึก
[ lambda1, b1, r1, app1, tau1 ] = [ 1, 1.5, 1, 0.0, 1 ]
[ lambda2, b2, r2, app2, tau2 ] = [ 1, -1.5, 1, 0.3, 1 ]
# initial conditions
x1_0 = 0.5
x2_0 = 0.5

RJ()
````
